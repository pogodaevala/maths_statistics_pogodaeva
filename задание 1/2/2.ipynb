```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def mu(x, k, mean):
    sum = np.sum((x-mean)**k)
    return sum/np.size(x)

def mode(x):
    unique, counts = np.unique(x, return_counts=True)
    max_count = np.max(counts)
    modes = unique[counts == max_count]
    return modes
```

```python
x = np.random.exponential(scale=1, size=25)
```

```python
median = x[12]
scope = x.max() - x.min()

mean = np.mean(x)
mu_3 = mu(x, 3, mean)
mu_2 = mu(x, 2, mean)
asymmetry = mu_3/mu_2**(3/2)

modes = mode(x)
```

```python
print(f"медиана = {median}\n размах = {scope}\n коэффициент ассиметрии = {asymmetry}\n моды :{modes if np.size(modes)<25 else 'всего по 1'}")
```
